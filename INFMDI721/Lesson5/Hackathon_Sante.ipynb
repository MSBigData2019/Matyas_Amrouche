```python
# Librairies utiles
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Pre-processing of the data

```python
# Densité des médecins (par 100000 habitants) par spécialité et par territoire (régions et départements)
skip = np.arange(1,17)
densite_medecin = pd.read_excel('/Users/matyasamrouche/Downloads/rpps-medecins18-tab7v3_68541554493006.xlsx' 
                                , encoding = "utf-8"
                               , skiprows = skip)
densite_medecin = densite_medecin.rename(columns={'ZONE INSCRIPT': 'Département'})
densite_medecin['Département'] = densite_medecin['Département'].str[:2]
densite_medecin = densite_medecin.set_index('Département')
densite_medecin.head(5)
```

```python
densite_medecin.columns.get_values()
```

```python
# Population par age/territoire en France 
col_interessantes = ['Département'
                    , 'Population en 2015 (princ)'
                    , 'Pop 0-2 ans en 2015 (princ)'
                    , 'Pop 3-5 ans en 2015 (princ)'
                    , 'Pop 6-10 ans en 2015 (princ)'
                    , 'Pop 11-17 ans en 2015 (princ)'
                    , 'Pop 18-24 ans en 2015 (princ)'
                    , 'Pop 25-39 ans en 2015 (princ)'
                    , 'Pop 40-54 ans en 2015 (princ)'
                    , 'Pop 55-64 ans en 2015 (princ)'
                    , 'Pop 65-79 ans en 2015 (princ)'
                    , 'Pop 80 ans ou plus en 2015 (princ)']


pop = pd.read_excel('/Users/matyasamrouche/Downloads/base-ic-evol-struct-pop-2015.xls'
                    , skiprows = [0,1,2,3,5] 
                    , encoding = "utf-8")

pop = pop[col_interessantes]
pop = pop.groupby(['Département']).sum()
pop.head(5)
```

```python
# Effectif des médecins  par spécialité et par territoire (régions et départements)
skip = np.arange(1,17)
densite_medecin = pd.read_excel('/Users/matyasamrouche/Downloads/rpps-medecins18-tab7v3_68541554493006.xlsx' 
                                , encoding = "utf-8"
                               , skiprows = skip)
densite_medecin = densite_medecin.rename(columns={'ZONE INSCRIPT': 'Département'})
densite_medecin['Département'] = densite_medecin['Département'].str[:2]
densite_medecin.head(5)
```

# 3) Corrélation entre densité de médecins (par spé) et densité de population (par âge) par département

Ici nous nous intéresserons à 3 populations :
    - les bébés
    - les personnes âgées (vieux)


```python
Specialites = ['Chirurgie infantile'
              , 'GÈriatrie'
              , 'PÈdiatrie']
```

```python
k = 0
fig = plt.figure(figsize=(16,4)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

enfants = pop['Pop 0-2 ans en 2015 (princ)'] + pop['Pop 3-5 ans en 2015 (princ)'] + pop['Pop 6-10 ans en 2015 (princ)']
densite_medecin[Specialites[k]].plot(kind='bar', color='blue', ax=ax, width=width, position=1, label='densité_pédiatre')
enfants.plot(kind='bar', color='red', ax=ax2, width=width, position=0, label='enfants')
ax.set_ylabel('Densité chirurgien infantile')
ax2.set_ylabel('Population enfants en bas âge')
plt.title(f'Spécialité : {Specialites[k]}')    
plt.legend()
plt.show()
```
